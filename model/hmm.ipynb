```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../dataset/combined_set.csv')
df.head(10)
```

```python

from hmmlearn import hmm

# Split the data into training (2018-2021) and testing (2022)
train_data = df[df['Sail Date.'] < '2022-01-01']
test_data = df[df['Sail Date.'] >= '2022-01-01']

# Extract the quantities for training
train_quantities = train_data['QTY.(KGS)'].values.reshape(-1, 1)

# Initialize and train the HMM
model = hmm.GaussianHMM(n_components=2, covariance_type="full", n_iter=100)
model.fit(train_quantities)

# Predict quantities for 2022 along with dates
test_dates = test_data['Sail Date.'].values.reshape(-1, 1)
predicted_data, _ = model.sample(len(test_data))
predicted_df = pd.DataFrame({
    'Sail Date.': pd.date_range(start='2022-01-01', periods=len(test_data)),
    'Predicted QTY.(KGS)': predicted_data.flatten()
})

# Print the predicted data along with dates
print(predicted_df)
```
