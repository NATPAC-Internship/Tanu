```python
import folium
import pandas as pd

df=pd.read_csv("./../../../dataset/prediction/LightGBM_chumma.csv")
kollam_df = df[df['District'] == "Kottayam"]
```

```python
#2024 plotting
columns_to_drop = ['2025', '2026', '2027', '2028']
filtered_2024_df = kollam_df.drop(columns=columns_to_drop, axis=1)

mean_latitude = filtered_2024_df['Latitude'].mean()
mean_longitude = filtered_2024_df['Longitude'].mean()
mymap = folium.Map(location=[mean_latitude, mean_longitude], zoom_start=5)

# Function to map quantity values to circle sizes (radius)
def get_circle_radius(amount):
    if amount < 100:
        return .5
    elif amount < 1000:
        return 1.0
    elif amount < 5000:
        return 1.5
    elif amount < 10000:
        return 2.0
    elif amount < 50000:
        return 2.5
    elif amount < 150000:
        return 3.0
    elif amount < 100000:
        return 3.5
    elif amount < 200000:
        return 4.0
    else:
        return 4.5


for index, row in filtered_2024_df.iterrows():
    
    amount = row['2024']  
    circle_radius = get_circle_radius(amount)
    
  
    folium.CircleMarker([row['Latitude'], row['Longitude']], 
                        radius=circle_radius, 
                        color='maroon',  
                        fill=True, 
                        fill_color='maroon',  
                        fill_opacity=0.6,
                        popup=f'Cluster {row["Cluster"]}, Amount: {amount}').add_to(mymap)

filename = f'2024_map_.html'
mymap.save(filename)
mymap
```
